```python
import wikipedia

def load_diseases(filename):
    diseases = []
    with open(filename, 'r') as file:
        for line in file:
            disease_name = line.strip()  
            diseases.append(disease_name)
    return diseases

diseases = load_diseases("common_diseases.txt")


```

```python

def scrape_disease_info(disease_name):
  try:
    # Attempt to fetch the Wikipedia page content
    page_content = wikipedia.page(f"disease {disease_name}").content
    # Create the filename with proper formatting
    filename = f"disease_text/{disease_name}.txt"

    # Write the content to the text file
    with open(filename, 'w', encoding='utf-8') as file:
      file.write(page_content)

    print(f"Successfully scraped information for {disease_name} and saved to {filename}")
    return True

  except wikipedia.exceptions.PageError:
    print(f"Error: Wikipedia page not found for {disease_name}")
    return False

  except Exception as e:
    print(f"An unexpected error occurred: {e}")
    return False
```

```python
for disease in diseases:
    disease = disease.replace('/', ' ')
    scrape_disease_info(disease)
    
```

```python
import os
import re

def replace_headings(text):
    # Regex patterns to match the different heading levels
    pattern_h4 = re.compile(r'==== (.*?) ====')
    pattern_h3 = re.compile(r'=== (.*?) ===')
    pattern_h2 = re.compile(r'== (.*?) ==')
    
    # Replace the headings with Markdown-style headings
    text = pattern_h4.sub(r'#### \1', text)
    text = pattern_h3.sub(r'### \1', text)
    text = pattern_h2.sub(r'## \1', text)
    
    return text

def update_files_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"): 
            file_path = os.path.join(folder_path, filename)
            
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
            
            new_content = replace_headings(content)
            
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(new_content)
            
            print(f"Updated: {filename}")

# Example usage
folder_path = '/Users/sudhanshurai/Desktop/Medical Diagnosis/Mediassist/disease_text'
update_files_in_folder(folder_path)
```
